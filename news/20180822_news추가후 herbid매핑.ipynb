## 엑셀 -> db로

```python
def xlsx_to_db(filename):
    pd.read_excel(filename+'.xlsx',sheet_name=None)
    con=sqlite3.connect(filename+".db")
    wb=pd.read_excel(filename+'.xlsx',sheet_name=None)
    for sheet in wb:
        wb[sheet].to_sql(sheet,con, index=False)
    con.commit()
    con.close()
```

```python
xlsx_to_db( filename )
```
